```python
import json
import csv
import tkinter as tk
from tkinter import ttk

with open("C:/Users/MSI-PC/Desktop/example_product_json.json", "r", encoding="utf-8") as file:
    data = json.load(file)

def analyze_data_types(data, parent_key="root"):
    types_info = {}
    
    def traverse(obj, key_prefix=""):
        if isinstance(obj, dict):
            for key, value in obj.items():
                new_key = f"{key_prefix}.{key}" if key_prefix else key
                traverse(value, new_key)
        elif isinstance(obj, list):
            types_info[key_prefix] = "list"
            for i, item in enumerate(obj[:3]): 
                traverse(item, f"{key_prefix}[{i}]")
        else:
            types_info[key_prefix] = type(obj).__name__
    
    traverse(data)
    return types_info

# Оценка полезности компонентов
def evaluate_features(data):
    useful_features = []
    non_useful_features = []
    
    for item in data.get("layout", []):
        component = item.get("component", "Unknown")
        # Полезные компоненты содержат ключевые слова
        if any(kw in component.lower() for kw in ["search", "price", "cart", "gallery"]):
            useful_features.append(component)
        else:
            non_useful_features.append(component)
    
    return useful_features, non_useful_features

# Функция для визуального отображения полезности компонентов
def display_feature_evaluation(data):
    root = tk.Tk()
    root.title("Анализ полезности фич")
    
    frame = ttk.Frame(root, padding=10)
    frame.pack(fill="both", expand=True)
    
    ttk.Label(frame, text="Полезные фичи", font=("Arial", 14, "bold"), foreground="black").pack()
    useful, non_useful = evaluate_features(data)
    for feature in useful:
        ttk.Label(frame, text=feature, font=("Arial", 12)).pack()
    
    ttk.Label(frame, text="Не очень полезные фичи", font=("Arial", 14, "bold"), foreground="red").pack()
    for feature in non_useful:
        ttk.Label(frame, text=feature, font=("Arial", 12)).pack()
    
    root.mainloop()

display_feature_evaluation(data)
```
